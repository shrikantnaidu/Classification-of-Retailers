```python
cols = list(df)
for column in cols:
    if column in df.select_dtypes(include=np.number).columns:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        fence_low = q1 - (1.5*iqr)
        fence_high = q3 + (1.5*iqr)
        print(column + ' ---------', df.loc[(df[column] < fence_low) | (df[column] > fence_high)].shape[0])
```

# Preparing data for Clustering

```python
df1 = df.copy()
df1.drop('Retailer Name',axis=1,inplace=True)
```

```python
df1.head()
```

# Why Normalization?
Higher no. will be taken as more weight which would disturb the balance of the already sparsed data.

```python
from sklearn.preprocessing import normalize
dfd = normalize(df1)
dfd = pd.DataFrame(dfd, columns=df1.columns)
```

```python
X = np.array(dfd)
from sklearn.cluster import KMeans
wcss =[]
for i in range (1,30):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans = kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
from matplotlib import pyplot as plt
plt.plot(range(1,30),wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')  #Within Cluster sum of squares
plt.show()
```

# Observation
The clusters start a dip between 0 and 5 so maybe it's a 2.5

```python
kmeans=KMeans(n_clusters = 3,init='k-means++',random_state = 0)

cluster_labels = kmeans.fit_predict(X)

clusters=pd.concat([df1, pd.DataFrame({'cluster':cluster_labels})], axis=1)
clusters.head()
```

```python
data['Category of Retailer'] = clusters['cluster']
data.head()
```

```python
import seaborn as sns
for c in clusters:
    grid= sns.FacetGrid(clusters, col='cluster')
    grid.map(plt.hist, c)
```

# Insights
1. cluster 1 : Regular customers with good sales and variety of orders(Good Customers)  
2. cluster 2 : Irregular customers  #Worst(Bad Customers) 
3. cluster 3 : Regular customers with better sales and less varied orders(Great Customers)

```python
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity

distance = 1 - cosine_similarity(X)

pca = PCA(2)
pca.fit(distance)
X_PCA = pca.transform(distance)
X_PCA.shape
```

```python
x, y = X_PCA[:, 0], X_PCA[:, 1]

colors = {0: 'red',
          1: 'green',
          2: 'black'}

names = {0: 'Good Customers' , 
         1: 'Bad Customers' ,
         2: 'Great Customers'}
df1 = pd.DataFrame({'x': x, 'y':y, 'label':cluster_labels}) 
groups = df1.groupby('label')

fig, ax = plt.subplots(figsize=(20, 13)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.legend()
ax.set_title("Classification of Retailers.")
plt.show()
```
